```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
from sklearn.utils import shuffle
from sklearn.tree import DecisionTreeClassifier
from sklearn.feature_selection import VarianceThreshold
from sklearn.externals.six import StringIO
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import export_graphviz
from IPython.display import Image  
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import re
import spotipy
from spotipy import util
from spotipy.oauth2 import SpotifyClientCredentials
```

Agora, como cada música, em muitos casos, têm mais de um gênero e como existem gêneros muito específicos, é necessário juntar alguns gêneros

```python
pd.options.mode.chained_assignment = None


data = pd.DataFrame(columns = ['artists','name','year','popularity','acousticness','danceability','energy','instrumentalness','liveness','speechiness','valence','tempo','duration_ms','genres'])
data = pd.read_csv('data_generos.csv')


for i,item in enumerate(data['genres']):
    if item != str:
        item = str(item)
    if "jazz" in item or "lounge" in item:
        data['genres'][i] = 'jazz'
    elif "rock" in item:
        data['genres'][i] = 'rock'
    elif "pop" in item:
        data['genres'][i] = 'pop'
    elif "metal" in item or "punk" in item:
        data['genres'][i] = 'punk/metal'
    elif "country" in item or "folk" in item:
        data['genres'][i] = 'country/folk'
    elif "funk" in item or "disco" in item or 'r&b' in item or 'motown' in item:
        data['genres'][i] = 'funk/disco/r&b'
    elif "edm" in item or "house" in item:
        data['genres'][i] = 'electronic'
    elif "hip hop" in item or 'trap' in item or 'rap' in item or 'reggaeton' in item:
        data['genres'][i] = 'hip hop'
    else:
        data['genres'][i] = 'NaN'
    
```

Agora já é possível fazer a Análise Exploratória Inicial:

```python
data_media = data
data_media = data_media.dropna(axis=0, subset=['genres'])
data_media = data.groupby('genres').mean()
data_media = data_media.drop('year',axis=1)
data_media = data_media.drop('NaN')
```

```python
# Normalizando para que as diferenças sejam mais perceptíveis
norm = data_media.apply(lambda x: x/x.max(), axis=0)
norm
```

```python
# Variancias
seletor = VarianceThreshold()
seletor.fit_transform(norm)
variancas = seletor.variances_


rowvar = pd.DataFrame(variancas.reshape(-1, len(variancas)), columns=norm.columns)

norm = norm.append(rowvar).rename({0: 'variancia'})
norm = norm.sort_values(by='variancia', ascending=False, axis=1)
norm
```

```python
X = data.drop(columns=['name', 'artists', 'genres',])
y = data.genres
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
dtc = DecisionTreeClassifier(random_state=42)
dtc.fit(X_train, y_train)
```

```python
dtc.score(X_test, y_test)
```
